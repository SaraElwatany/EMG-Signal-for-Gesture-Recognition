```python
# import required modules
import pandas as pd
import numpy as np
import os




# Assign directory
directory = "data"
csv_directory = directory + "/csv files"





# Iterate over each txt file in that directory & convert to csv file & save in (csv files) folder
def convert_txt_to_csv():

    count = 1

    for file in os.scandir(directory):
            
        # Check if it is a file, if so proceed
        if file.is_file():

            #print(filename.path)
            data = pd.read_csv(file.path, sep='\s+', engine="python")       
            path = csv_directory + "/" + str(count) + ".csv"
            #print(data, path)
            data.to_csv(path, index = None)
            count+=1
    





# Iterate over each csv file in that directory & add label column
def add_lbl_to_csv():

    for file in os.scandir(csv_directory):

        # Check if it is a file, if so proceed
        if file.is_file():

            if (file.name != "merged_files.csv"):

                data_new = pd.read_csv(file.path, engine="python")           # Read CSV file
                name = file.name.split(".", 1)
                subject = int(name[0])
                #print(subject)
                if ((subject%2) == 0):
                    data_new["label"] = subject - (subject / 2)
                else:
                    subject+=1
                    data_new["label"] = subject - (subject / 2)                      # Append list as new variable 
                                            
                data_new.to_csv(file.path, index = None)        # Write DataFrame to csv file, but with new header  





# Remove Records of unknown class
def remove_unknown():
        
    for file in os.scandir(csv_directory):

        # Check if it is a file, if so proceed
        if file.is_file():

            if (file.name != "merged_files.csv"):
                data= pd.read_csv(file.path, engine="python")           # Read CSV file
                data.drop(data[data['class'] == 0].index, inplace = True)                                           
                data.to_csv(file.path, index = None)        # Write DataFrame to csv file, but with new header  






# Merge all csv files to one csv file & save in (csv files/merged_files.csv) file
def merge_csv():

    df = pd.DataFrame()

    for file in os.scandir(csv_directory):
            
        # Check if it is a file, if so proceed
        if file.is_file():

            data = pd.read_csv(file)
            df = pd.concat([df, data], axis=0)

    path = csv_directory + "/merged_files.csv" 
    df.to_csv(path, index = None)






```

```python
convert_txt_to_csv()
```

```python
add_lbl_to_csv()
```

```python
remove_unknown()
```

```python
merge_csv()
```

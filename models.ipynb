```python
import pandas as pd
import numpy as np
```

```python
features_path = 'features.csv'
df = pd.read_csv(features_path)
```

```python
df
```

```python
df = df.drop(0)
```

```python
df
```

```python
df=df.drop(columns=['Unnamed: 0'])
```

```python
df.columns
```

```python
# Separate the features (X) and the target variable (Y)
X = df.drop(['class','label'],axis=1).values  
Y = df['class'].values
```

```python
X
```

```python
Y
```

```python
from sklearn.feature_selection import SelectKBest, f_classif
import numpy as np

# Feature selection with SelectKBest and f_classif
k = 65  # Number of features to select
feature_selector = SelectKBest(score_func=f_classif, k=k)
fit = feature_selector.fit(X, Y)

# Summarize scores
feature_scores = fit.scores_
print("Feature Scores:",feature_scores.shape)
print(feature_scores)

# Rank features based on scores
ranked_indices = np.argsort(feature_scores)[::-1]
ranked_features = X[:, ranked_indices]

# Select subset of features
selected_features = ranked_features[:, :k]
```

```python
print(ranked_indices.shape)
ranked_indices
```

```python
print(ranked_features.shape)
ranked_features
```

```python
print(selected_features.shape)
selected_features
```

```python
# Visualization 
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.bar(range(len(feature_scores)), feature_scores)
plt.xlabel("Feature Index")
plt.ylabel("Feature Score")
plt.title("Feature Scores")
plt.show()
```

```python
from sklearn.model_selection import train_test_split

# Split the data into training and testing sets
X_train, X_test, Y_train, Y_test = train_test_split(selected_features, Y, test_size=0.2, random_state=42)
```

```python
from sklearn.svm import SVC

# Train the SVM classifier
svm = SVC()
svm.fit(X_train, Y_train)

# Evaluate the SVM classifier
svm_accuracy = svm.score(X_test, Y_test)
print("Accuracy:", svm_accuracy)
```

```python
from sklearn.ensemble import RandomForestClassifier

# Train the Random Forest classifier
rf = RandomForestClassifier()
rf.fit(X_train, Y_train)

# Evaluate the Random Forest classifier
rf_accuracy = rf.score(X_test, Y_test)
print("Accuracy:", rf_accuracy)
```

```python
from sklearn.neural_network import MLPClassifier
mlp = MLPClassifier(hidden_layer_sizes=(10,15,20), activation='relu',
solver='adam', max_iter=500)
mlp.fit(X_train,Y_train)
predict_train = mlp.predict(X_train)
predict_test = mlp.predict(X_test)

# Evaluate the Random Forest classifier
mlp_accuracy = mlp.score(X_test, Y_test)
print("Accuracy:", mlp_accuracy)
```

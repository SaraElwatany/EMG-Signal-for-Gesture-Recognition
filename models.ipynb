```python
import pandas as pd
import numpy as np
```

```python
features_path = 'features.csv'
df = pd.read_csv(features_path)
```

```python
df
```

```python
df = df.drop(0)
```

```python
df
```

```python
df=df.drop(columns=['Unnamed: 0'])
```

```python
df.columns
```

```python
# Separate the features (X) and the target variable (Y)
X = df.drop(['class','label'],axis=1).values  
Y = df['class'].values
```

```python
X
```

```python
Y
```

```python
from sklearn.feature_selection import SelectKBest, f_classif
import numpy as np

# Feature selection with SelectKBest and f_classif
k = 65  # Number of features to select
feature_selector = SelectKBest(score_func=f_classif, k=k)
fit = feature_selector.fit(X, Y)

# Summarize scores
feature_scores = fit.scores_
print("Feature Scores:",feature_scores.shape)
print(feature_scores)

# Rank features based on scores
ranked_indices = np.argsort(feature_scores)[::-1]
ranked_features = X[:, ranked_indices]

# Select subset of features
selected_features = ranked_features[:, :k]
```

```python
print(ranked_indices.shape)
ranked_indices
```

```python
print(ranked_features.shape)
ranked_features
```

```python
print(selected_features.shape)
selected_features
```

```python
# Visualization 
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.bar(range(len(feature_scores)), feature_scores)
plt.xlabel("Feature Index")
plt.ylabel("Feature Score")
plt.title("Feature Scores")
plt.show()
```

```python
from sklearn.model_selection import train_test_split

# Split the data into training and testing sets
X_train, X_test, Y_train, Y_test = train_test_split(selected_features, Y, test_size=0.2, random_state=42)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, cohen_kappa_score, mean_squared_error, r2_score
import numpy as np
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
import seaborn as sns
```

```python
from sklearn.svm import SVC
from sklearn import metrics

# Train the SVM classifier
svm = SVC()
svm.fit(X_train, Y_train)
Y_pred = svm.predict(X_test)


# calculate the evaluation metrics
accuracy = accuracy_score(Y_test, Y_pred)
precision = precision_score(Y_test, Y_pred, average='weighted')
recall = recall_score(Y_test, Y_pred, average='weighted')
f1 = f1_score(Y_test, Y_pred, average='weighted')
rmse = np.sqrt(mean_squared_error(Y_test, Y_pred))


# print the evaluation metrics
print("Accuracy:", accuracy)
print("Precision:", precision)
print("Recall:", recall)
print("F1 score:", f1)
print("Root Mean Squared Error:", rmse)


class_names = ['1', '2', '3', '4','5','6']

cf_matrix = confusion_matrix(Y_pred,Y_test)
sns.heatmap(cf_matrix, annot=True, xticklabels=class_names, yticklabels=class_names)

print(classification_report(Y_test, Y_pred))


```
